```python
'''A) STORING ALL IDs '''
student_ids=('s101','s102','s103','s104')
```

```python
'''B) STORING STUDENT RECORDS '''
students={
    's101':{'name':'asha','assignment':78,'test':80,'attendance':92,'hours':8},
    's102':{'name':'ravi','assignment':65,'test':68,'attendance':85,'hours':5},
    's103':{'name':'Meena','assignment':88,'test':90,'attendance':96,'hours':10},
    's104':{'name':'Kiran','assignment':55,'test':58,'attendance':78,'hours':4}
}
```

```python
'''C) CALCULATING AVERAGE'''
def calculate_average(sid):
    s_assignment_mark=students[sid]['assignment']
    s_assignment_test=students[sid]['test']
    s_score=s_assignment_mark+s_assignment_test
    s_average=(s_score)/2
    
    return s_average
```

```python
average=(calculate_average('s101'))
```

```python
'''D) DETERMINING RISK LEVEL '''
def determine_risk(sid):
    average=calculate_average(sid)
    if average>=75 and students[sid]['attendance']>=90 and students[sid]['hours']>=7:
        return "LOW RISK"
    elif average>=60 and students[sid]['attendance']>=80 and students[sid]['hours']>=5:
        return "MODERATE RISK"
    else:
        return "HIGH RISK"
```

```python
risk=determine_risk('s104')
```

```python
print(risk)
```

```python
'''E) PROCESS MULTIPLE STUDENTS RECORDS'''
for i in student_ids:
    if i in students.keys():
        print(students[i])
```

```python
'''F) DISPLAY PERFORMANCE REPORT'''
def get_student_report(sid):
    if sid not in students:
        print("Invalid Student ID")
        return 
    data=students[sid]
    avg=calculate_average(sid)
    risk=determine_risk(sid)
    
    print("STUDENT PERFORMANCE REPORT")
    print("-"*40)
    print("Name:",data['name'])
    print("Test Score:",data['test'])
    print("Assignement Score:",data['assignment'])
    print("Average score:",avg)
    print("Attendance(%):",data['assignment'])
    print("Study hours/week:",data['hours'])
    print("Risk Level:",risk)
    print("-"*40)
```

```python
get_student_report('s101')
```
